Esse notebook tem como único propósito servir de apoio para gerar os gráficos utilizados no post "Pare agora de usar Gráficos de Pizza" no medium do Paulo Vilarinho.

```
import numpy as np
import matplotlib.pyplot as plt
```

```
labels = ["Gosto Muito","Gosto","Irrelevante", "Não Gosto","Odeio"]
size = [25,30,10,15,20]

fig,ax = plt.subplots(figsize=(10,6))



ax.pie(size,labels=labels)
ax.legend(loc=1,bbox_to_anchor=(1.35, 1, 0, 0))
ax.set_title("Opinião sobre o novo produto")

fig.show()

fig.savefig("ex1")
```

```
fig,ax = plt.subplots(figsize=(10,6))



ax.bar([1,2,3,4,5],size,tick_label=labels)
ax.set_title("Opinião sobre o novo produto")

fig.show()

fig.savefig("ex2")
```

```


labels = ["Não", "Tambem não"]
size = [70,30]

fig,ax = plt.subplots(figsize=(10,6))



ax.pie(size,labels=labels)
ax.legend(loc=1,bbox_to_anchor=(1.2, 1, 0, 0))
ax.set_title("Graficos de pizza são bons?")

fig.show()

fig.savefig("meme")
```
